```python
# Now select the columns you want

# FIRST - Load your data into a DataFrame

df = pd.read_csv('tmdb_5000_movies.csv')  # or whatever your file is called

# NOW you can select columns
df = df[['title', 'overview', 'genres', 'vote_average', 'tagline','popularity']]


print(df.head())
```

```python
#filling null with empty space
df['overview']=df['overview'].fillna(' ')
```

```python
import ast
df['genres']=df['genres'].apply(lambda x: " ".join([i['name']for i in ast.literal_eval(x)]) )
```

```python
#filling null with empty space
df['tagline']=df['tagline'].fillna(' ')
```

```python
df['tags']=df['overview']+ " "+ df["genres"]+" " +df['tagline']
```

```python
#moving to nlp processing 
import nltk 
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re 
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    # Convert to lowercase
    text = str(text).lower()
    
    # Remove punctuation (keep only letters, numbers, and spaces)
    text = re.sub(r'[^\w\s]', ' ', text)
    
    # Split into words
    words = text.split()
    
    # Remove stopwords
    words = [word for word in words if word not in stop_words]  # Fixed: 'word' not 'words'
    
    # Lemmatize
    words = [lemmatizer.lemmatize(word) for word in words]
    
    return " ".join(words)
```

```python

df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df['tags'][0]
```

```python
df=df.reset_index(drop=True)
```

```python

indices = pd.Series(df.index, index=df['title']).drop_duplicates() 
indices
```

```python
#convert it into vector 
#apply the vectorization 
from sklearn.feature_extraction.text import TfidfVectorizer 
```

```python
tfidf=TfidfVectorizer(max_features=50000,ngram_range=(1, 2),stop_words='english')
```

```python
tfidf_matrix=tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

```python
#cosine similarity to predicting the movies 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n=10):
    if title not in indices:
      return ['Movie not found']
    idx= indices[title]
    sim_score=cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
    similar_idx=sim_score.argsort()[::-1][1:n+1]
    return df['title'].iloc[similar_idx]
    
```

```python
recommend('Toy Story')
```

```python
recommend('Avatar',4)
```

```python
import pickle 
pickle.dump(tfidf_matrix,open('tfidf_matrix.pkl','wb'))
pickle.dump(df,open('indices.pkl','wb'))
df.to_pickle('df.pkl')
pickle.dump(tfidf,open('tfidf.pkl','wb'))
```
